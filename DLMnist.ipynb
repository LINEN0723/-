```python
from keras.datasets.mnist import load_data
(x_train,y_train),(x_test,y_test)=load_data()
```

```python
print(x_train.shape)
print(y_train.shape)#訓練資料
print(x_test.shape)
print(y_test.shape)
```

```python
xx_train=x_train.reshape(60000,784)/255
xx_test=x_test.reshape(10000,784)/255
```

```python
xx_train[0]
```

```python
from keras.utils.np_utils import to_categorical
yy_train = to_categorical(y_train)
yy_test = to_categorical(y_test)
```

```python
yy_train[0]
```

```python
from keras.models import Sequential
model = Sequential()
```

```python
from keras.layers import Dense
model.add(Dense(128,activation="relu",input_dim=784))
model.summary()#784*128+128=100480
```

```python
model.add(Dense(10,activation='softmax'))
model.summary()#summary可以看model
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
#[訓練模型]—選擇[batch_size]告訴keras一次訓練(權重調整)的資料量有多少,太大與太小都不宜,一般約設置在[100~200]之間
```

```python
train_history=model.fit(xx_train,yy_train,batch_size=100,epochs=10,validation_split=0.15)#訓練10組
#1~5欠擬和
#6~8完美擬和
#9~10過擬和
```

```python
train_history=model.fit(xx_train,yy_train,batch_size=100,epochs=20,validation_split=0.15)#訓練10組
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(train_history.history['loss'])  
plt.plot(train_history.history['val_loss'])  
plt.title('Train History')  
plt.ylabel('loss')  
plt.xlabel('Epoch')  
plt.legend(['loss', 'val_loss'], loc='upper left')  
plt.show() 
```

```python
#評估模型
model.evaluate(xx_test,yy_test)
```

```python
#預測機率
model.predict(xx_test)
```

```python
#預測答案
#.argmax預測最大的丟出來
import numpy
pre = model.predict(xx_test)
pre=numpy.argmax(pre,axis=1)
pre
```

```python
#查看預測
y_test
```

```python
from sklearn.metrics import confusion_matrix
import pandas
#panda混淆矩陣近來正確答案跟預測的
pandas.DataFrame(confusion_matrix(y_test,pre))

```

```python
#找錯誤的資料的比數
numpy.nonzero(pre != y_test)
false_index=numpy.nonzero(pre != y_test)[0]
print("錯誤次數: ",len(false_index))
plt.imshow(x_test[115],cmap="gray")
false_img = x_test[false_index]
ori_label = y_test[false_index]
pre_label = pre[false_index]
cols = 10
rows = int(len(false_index) / 10) + 1
plt.figure(figsize=(14, 40))
for (index, img) in enumerate(false_img):
    plt.subplot(rows, cols, index + 1)
    msg = "[O]:" + str(ori_label[index]) + "[P]:" + str(pre_label[index])
    plt.title(msg)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
```
